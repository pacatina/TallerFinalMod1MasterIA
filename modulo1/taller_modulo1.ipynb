### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí
promedio_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]

#Variables 
sum = 0
max = promedio_vida [0]
min = promedio_vida [0]

#Recorre la lista
for cant in promedio_vida:
   sum += cant
   
   if cant > max:
    max = cant
    
   elif cant < min:
    min = cant

    
# Promedio
promedio = sum / len(promedio_vida)
print(f"{promedio:.2f}")

# Máximo
print(f"El promedio Maximo es : {max:.2f}")

# Mínimo
print(f"El promedio Maximo es : {min:.2f}")
````

### 1.4 — Listas + Funciones + Bucles (integrador)

**Ejercicio 1.4** — Usa la función creada en 1.3 y la lista del ejercicio 1.1 para:
1. Crear una **nueva lista** llamada `clasificaciones` que contenga la clasificación de cada esperanza de vida
2. Imprimir cada par: `(valor, clasificación)` usando un bucle `for`

````python
# Tu código aquí
#Lista punto 1.1
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]

#Creacion Nueva Lista
clasificaciones = []

#Funcion creada en punto 1.3
def clasificar_esperanza_vida(num):
            
    if num >= 75:
        return("Alta")
    elif num >=60 and num <= 74.9:
        return("Media") 
    else:
        return("Baja")

#Creacion de lista de clasificaciones
for valor in esperanzas_de_vida:
    clasificacion = clasificar_esperanza_vida(valor)
    clasificaciones.append((valor, clasificacion))
#for para impresion de los pares
for valor, clasificacion in clasificaciones:
    print(valor, clasificacion)
    
````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Tu código aquí

#Creamos un nuevo archivo para dejar el original como estaba inicialmente

salud_nuevo = salud.drop_duplicates()

#Creamos las medias de cada una de las columnas con nulos
media_esperanza_vida = salud_nuevo["esperanza_vida"].mean()
print(f"Media de esperanza de vida: {media_esperanza_vida:.2f}")
salud_nuevo["esperanza_vida"] = salud_nuevo["esperanza_vida"].fillna(media_esperanza_vida)

media_gasto_salud = salud_nuevo["gasto_salud_usd"].mean()
print(f"Media de gasto de salud USD: {media_gasto_salud:.2f}")
salud_nuevo["gasto_salud_usd"] = salud_nuevo["gasto_salud_usd"].fillna(media_gasto_salud)

media_mortalidad_infantil = salud_nuevo["mortalidad_infantil"].mean()
print(f"Media de mortalidad infantil: {media_mortalidad_infantil:.2f}")
salud_nuevo["mortalidad_infantil"] = salud_nuevo["mortalidad_infantil"].fillna(media_mortalidad_infantil)

media_medicos = salud_nuevo["medicos_por_1000"].mean()
print(f"Media de médicos por 1000: {media_medicos:.2f}")
salud_nuevo["medicos_por_1000"] = salud_nuevo["medicos_por_1000"].fillna(media_medicos)

media_obesidad = salud_nuevo["obesidad_pct"].mean()
print(f"Media de obesidad PCT: {media_obesidad:.2f}")
salud_nuevo["obesidad_pct"] = salud_nuevo["obesidad_pct"].fillna(media_obesidad)

media_diabetes = salud_nuevo["diabetes_pct"].mean()
print(f"Media de diabetes PCT: {media_obesidad:.2f}")
salud_nuevo["diabetes_pct"] = salud_nuevo["diabetes_pct"].fillna(media_diabetes)

media_tabaquismo = salud_nuevo["tabaquismo_pct"].mean()
print(f"Media de tabaquismo PCT: {media_tabaquismo:.2f}")
salud_nuevo["tabaquismo_pct"] = salud_nuevo["tabaquismo_pct"].fillna(media_tabaquismo)

media_camas_hospital = salud_nuevo["camas_hospital_por_1000"].mean()
print(f"Media de camas hospital por 1000: {media_camas_hospital:.2f}")
salud_nuevo["camas_hospital_por_1000"] = salud_nuevo["camas_hospital_por_1000"].fillna(media_camas_hospital)

media_vacunacion = salud_nuevo["vacunacion_pct"].mean()
print(f"Media de vacunación PCT: {media_vacunacion:.2f}")
salud_nuevo["vacunacion_pct"] = salud_nuevo["vacunacion_pct"].fillna(media_vacunacion)
print(f"Filas y Columnas despues de limpiar: {salud_nuevo.shape} ")
salud_nuevo.isnull().sum()
````
